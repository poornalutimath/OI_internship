```python
#Basic BMI Calculator
def calculate_bmi(weight, height):
    return weight / (height ** 2)


def classify_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"


try:
    weight = float(input("Enter your weight in kilograms: "))
    height = float(input("Enter your height in meters: "))
    
    if weight > 0 and height > 0:
        bmi = calculate_bmi(weight, height)
        category = classify_bmi(bmi)
        print(f"Your BMI is {bmi:.2f}. You are classified as {category}.")
    else:
        print("Please enter positive values for weight and height.")
except ValueError:
    print("Invalid input. Please enter numbers.")


```

```python
import tkinter as tk
from tkinter import messagebox
import sqlite3

# Database setup 
conn = sqlite3.connect(':memory:')  
c = conn.cursor()
c.execute('''CREATE TABLE IF NOT EXISTS bmi_records
             (id INTEGER PRIMARY KEY, name TEXT, weight REAL, height REAL, bmi REAL)''')
conn.commit()

def calculate_bmi(weight, height):
    return weight / (height ** 2)

def classify_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

def save_data(name, weight, height, bmi):
    c.execute("INSERT INTO bmi_records (name, weight, height, bmi) VALUES (?, ?, ?, ?)",
              (name, weight, height, bmi))
    conn.commit()

def submit():
    try:
        name = entry_name.get()
        weight = float(entry_weight.get())
        height = float(entry_height.get())
        
        if weight > 0 and height > 0:
            bmi = calculate_bmi(weight, height)
            category = classify_bmi(bmi)
            save_data(name, weight, height, bmi)
            messagebox.showinfo("Result", f"Your BMI is {bmi:.2f}. You are classified as {category}.")
        else:
            messagebox.showerror("Error", "Please enter positive values.")
    except ValueError:
        messagebox.showerror("Error", "Invalid input. Please enter numbers.")

def create_gui():
    window = tk.Tk()
    window.title("BMI Calculator")

    global entry_name, entry_weight, entry_height
    
    tk.Label(window, text="Name:").grid(row=0, column=0)
    tk.Label(window, text="Weight (kg):").grid(row=1, column=0)
    tk.Label(window, text="Height (m):").grid(row=2, column=0)

    entry_name = tk.Entry(window)
    entry_weight = tk.Entry(window)
    entry_height = tk.Entry(window)

    entry_name.grid(row=0, column=1)
    entry_weight.grid(row=1, column=1)
    entry_height.grid(row=2, column=1)

    tk.Button(window, text="Submit", command=submit).grid(row=3, column=1)

    window.mainloop()

if __name__ == "__main__":
    create_gui()
```
